```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Iris.csv")
df.head()
print(df.shape)
df.columns
df.describe()
df.groupby("Species").mean()

plt.figure()
plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"]
)

plt.hist(df["SepalWidthCm"], bins=10)
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Width")
plt.show()
```
